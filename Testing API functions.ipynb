```python
def get_artist_songs(artist_id, pagination, headers):
    """
    Funciton used to get the songs for a specific artist.
    
    The first page is page=1, opposed to page=0
    """
    
    # URL for the artist's songs
    artists_songs = base_url + "/artists/" + artist_id + "/songs"
    
    # Request parameters
    params = {
        "per_page":50,
        "sort":"popularity",
        "page":pagination
    }
    
    response = requests.get(artists_songs, headers=headers, params=params)
    data_as_json = response.json()
    if data_as_json["meta"]["status"] == 200:
        return data_as_json["response"]["songs"]
```

```python
first_batch = get_artist_songs(artist_id, 1, headers)
first_batch
```

```python
second_batch = get_artist_songs(artist_id, 2, headers)
```

```python
def get_all_songs_by_artist(artist_id, headers):
    all_songs = []
    current_request = [True]
    current_page = 1
    while len(current_request)>0:
        print(f"on page {current_page}. total songs collected", end="\r")
        current_request = get_artist_songs(artist_id, current_page, headers)
        all_songs.extend(current_request)
        current_page +=1
        time.sleep(.25)
    return all_songs
```

```python
kendricks_songs = get_all_songs_by_artist(artist_id, headers)
```
